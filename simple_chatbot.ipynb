# ChatBot:🤖

# Importing the necessary python libraries

```python
from nltk.chat.util import Chat, reflections
```

## creating a list of queries and their responses

```python
conv = [

    # --- Greetings ---
    [
        r"hi|hello|hey",
        ["Hello! How can I help you today?"]
    ],
    [
        r"good morning",
        ["Good morning! Hope your day starts great."]
    ],
    [
        r"good afternoon",
        ["Good afternoon! What can I do for you?"]
    ],
    [
        r"good evening",
        ["Good evening! How can I assist you tonight?"]
    ],

    # --- About Chatbot ---
    [
        r"(.*)your name",
        ["My name is Klaus, I’m your chatbot assistant."]
    ],
    [
        r"who are you",
        ["I am Klaus, your simple chatbot friend."]
    ],
    [
        r"what can you do",
        ["I can chat with you, answer questions, and keep you company!"]
    ],

    # --- User Asking About Themselves ---
    [
        r"(.*)my name",
        ["I’m not sure, you haven’t told me your name yet!"]
    ],
    [
        r"i am (.*)",
        ["Nice to meet you, %1!"]
    ],
    [
        r"my name is (.*)",
        ["Hello %1! How can I help you today?"]
    ],

    # --- Feelings / Emotions ---
    [
        r"i am sad|i feel sad",
        ["I’m sorry to hear that. Want to talk about it?"]
    ],
    [
        r"i am happy|i feel happy",
        ["That's awesome! I’m happy for you 😊"]
    ],
    [
        r"i am bored",
        ["Maybe we can chat to pass the time!"]
    ],

    # --- Small Talk ---
    [
        r"how are you",
        ["I'm doing great, thanks for asking! How about you?"]
    ],
    [
        r"what are you doing",
        ["Just waiting here to chat with you!"]
    ],
    [
        r"where are you from",
        ["I live inside your computer 😄"]
    ],
    [
        r"do you like (.*)",
        ["I’m a bot, but I think %1 is interesting!"]
    ],

    # --- Jokes / Fun ---
    [
        r"tell me a joke",
        ["Why don’t robots panic? Because they have nerves of steel!"]
    ],
    [
        r"tell me something funny",
        ["Sometimes I talk to myself… because I’m the only one who listens perfectly!"]
    ],

    # --- Knowledge ---
    [
        r"what is (.*)",
        ["Hmm… I’m not fully sure about %1, but I can try to learn!"]
    ],
    [
        r"who is (.*)",
        ["I might not know exactly who %1 is, but I can guess they’re interesting!"]
    ],

    # --- Thanks / Closing ---
    [
        r"thank you|thanks",
        ["You’re welcome! Happy to help 😊"]
    ],
    [
        r"bye|goodbye|see you",
        ["Goodbye! Have a great day!"]
    ],
]
```

```python
print("Hi, I'm your assistance, you can ask me any question")
chat=Chat(conv, reflections)
chat.converse()
```
